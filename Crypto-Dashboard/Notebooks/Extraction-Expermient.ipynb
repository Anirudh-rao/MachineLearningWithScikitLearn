# API Ingestion From API Key

```python
import os
from dotenv import load_dotenv

load_dotenv() # Parses the .env file and loads it into the environment

AlphaAdvantageApi = os.getenv('AlphaAdvantage')
```

```python
url = f'https://www.alphavantage.co/query?function=CURRENCY_EXCHANGE_RATE&from_currency=BTC&to_currency=USD&apikey={AlphaAdvantageApi}'
r = requests.get(url)
data = r.json()
```

```python
crypto_data = pd.DataFrame(data)
crypto_data.head(100)
```

# Daily Data


```python
url = f'https://www.alphavantage.co/query?function=DIGITAL_CURRENCY_DAILY&symbol=BTC&market=EUR&apikey={AlphaAdvantageApi}'
r = requests.get(url)
data = r.json()

crypto_data = pd.DataFrame(data)
crypto_data.head(100)
print(crypto_data.shape)
```

```python
crypto_data
```

```python
crypto_data = pd.json_normalize(data)
crypto_data
```
